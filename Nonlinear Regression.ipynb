```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define a nonlinear function (e.g., exponential decay)
def nonlinear_func(x, a, b, c):
    return a * np.exp(-b * x) + c

# Generate synthetic data
np.random.seed(42)
x_data = np.linspace(0, 4, 50)  # Independent variable
y_data = nonlinear_func(x_data, 2.5, 1.3, 0.5) + np.random.normal(scale=0.2, size=len(x_data))  # Dependent variable with noise

# Perform nonlinear regression
try:
    # Fit the curve
    popt, pcov = curve_fit(nonlinear_func, x_data, y_data, p0=[1, 1, 1])  # Initial guesses for parameters

    # Extract optimized parameters
    a_opt, b_opt, c_opt = popt

    print("Optimized Parameters:")
    print(f"a = {a_opt}, b = {b_opt}, c = {c_opt}")

    # Predict values
    y_pred = nonlinear_func(x_data, *popt)

    # Plot data and fitted curve
    plt.scatter(x_data, y_data, label='Data', color='blue')
    plt.plot(x_data, y_pred, label='Fitted Curve', color='red')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Nonlinear Regression Example')
    plt.show()

except Exception as e:
    print(f"An error occurred: {str(e)}")
```
